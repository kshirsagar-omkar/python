## Matplotlib Functions 

```python
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([1, 3, 5, 7])
plt.plot(x, y)
plt.show()
```

```python
x_axis = np.array([0, 3])
y_axis = np.array([2, 4])
plt.plot(x_axis, y_axis, '*')
plt.show()
```

```python
x_axis=np.array([4, 8, 6, 2])
y_axis=np.array([1, 5, 3, 7])
plt.plot(x_axis, y_axis)
plt.show()
```

```python
y_axis=np.array([2, 7, 1, 8, 3, 9, 4, 5])
plt.plot(y_axis)
plt.show()
```

```python
y_axis=np.array([3,6,9,12])
plt.plot(y_axis,marker='d')  # *,o,+,^,x,_,>,<,d,D,|
plt.show()
```

```python
y_axis = np.array([2, 7, 1, 8, 3, 9, 4, 5])
plt.plot(y_axis,linestyle='dashdot') # dashed, solid, dotted
plt.show()
```

```python
y_axis = np.array([2, 7, 1, 8, 3, 9, 4, 5])
plt.plot(y_axis, color='olive') 
plt.show()
```

```python
y_axis = np.array([2, 7, 1, 8, 3, 9, 4, 5])
plt.plot(y_axis,linewidth='5')
plt.show()
```

```python
x=np.array([4, 8, 6, 2])
y=np.array([1, 5, 3, 7])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x=np.array([1, 2, 3, 4, 5 ,6])
y=np.array([101.5, 137.9, 150.45, 90.85, 145.05, 160.13])
plt.plot(x,y)
plt.xlabel("Ball")
plt.ylabel("Speed")
plt.show()
```

```python
x=np.array([1, 2, 3, 4, 5 ,6])
y=np.array([101.5, 137.9, 150.45, 90.85, 145.05, 160.13])
plt.plot(x,y)
plt.title("Bowler Speed Chart") # loc='right/left/center'
plt.xlabel("Ball")
plt.ylabel("Speed")
plt.show()
```

```python
x=np.array([1, 2, 3, 4, 5 ,6])
y=np.array([101.5, 137.9, 150.45, 90.85, 145.05, 160.13])
plt.plot(x,y)
plt.title("Bowler Speed Chart") 
plt.xlabel("Ball")
plt.ylabel("Speed")
plt.grid() # axis='x/y'
plt.show()
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([1, 3, 5, 7])
plt.subplot(1,2,2)
plt.plot(x, y)
```

```python
# plot 1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.plot(x,y)

# plot 2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,3,2)
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y) 
plt.show()
```

```python
# plot-1
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y,color='lightgreen')

# plot-2
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y,color='yellow')
plt.show()
```

```python
x=np.array(["X", "Y", "Z"])
y=np.array([3, 7, 1])
plt.bar(x,y)      # for horizontal bar use- barh
plt.show()
```

```python
y=np.array([38,23,27,12])
plt.pie(y)
plt.show() 
```

```python
y=np.array([38,23,27,12])
mylabels=['W','X','Y','Z']
plt.pie(y,labels=mylabels)
plt.show()
```

```python
y=np.array([38,23,27,12])
mylabels=['W','X','Y','Z']
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
y=np.array([38,23,27,12])
mylabels=['W','X','Y','Z']
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode, shadow=True)
plt.show()
```

```python
y=np.array([38,23,27,12])
mylabels=['W','X','Y','Z']
mycolors=['lightgreen','lightblue','black','yellow']
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

```python
y=np.array([38,23,27,12])
mylabels=['W','X','Y','Z']
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

```python
y=np.array([38,23,27,12])
mylabels=['W','X','Y','Z']
plt.pie(y,labels=mylabels)
plt.legend(title="Parameter") #loc='left','right'
plt.show()
```
