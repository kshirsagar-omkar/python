# Seaborn Functions:

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('mpg')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.scatterplot(x='mpg',y='horsepower',data=df)
```

```python
sns.scatterplot(x='mpg',y='horsepower',data=df,hue='origin')
```

```python
sns.scatterplot(x='mpg',y='horsepower',data=df,hue='origin',palette=["black","cyan","lime"])
```

```python
sns.scatterplot(x='mpg',y='horsepower',data=df,hue='origin',palette=["black","cyan","lime"],size='origin')
```

```python
sns.histplot(df['mpg'],bins=10) 
```

```python
sns.histplot(df['mpg'],bins=10, kde=True)
```

```python
sns.barplot(x=df['cylinders'],y=df['horsepower'],hue=df['cylinders'],palette=['cyan','yellow','lime','crimson','silver'])
```

```python
sns.boxplot(x=df['cylinders'],y=df['horsepower'],hue=df['origin'],palette=['cyan','yellow','lime'])
```

```python
sns.violinplot(x=df['cylinders'],y=df['horsepower'],hue=df['origin'],palette=['cyan','yellow','lime'])
```

```python
sns.stripplot(x=df['cylinders'],y=df['horsepower'],hue=df['origin'],palette=['cyan','yellow','lime'])
```

```python
sns.stripplot(x=df['cylinders'],y=df['horsepower'],hue=df['origin'],palette=['cyan','yellow','lime'], dodge=True)
```

```python
sns.jointplot(x=df['cylinders'],y=df['horsepower'],hue=df['origin'],palette=['cyan','yellow','lime'], kind='kde')
```

```python
sns.jointplot(x=df['cylinders'],y=df['horsepower'],hue=df['origin'],palette=['cyan','yellow','lime'], kind='kde', fill=True)
```

```python
sns.jointplot(x=df['cylinders'],y=df['horsepower'],palette=['cyan','yellow','lime'], kind='hex', cmap='coolwarm')
```

```python
sns.pairplot(df[["mpg","cylinders","horsepower","origin"]],hue='cylinders')
```
